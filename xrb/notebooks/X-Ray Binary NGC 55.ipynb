```python
import os
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d

%matplotlib inline
```

### First, let's load up some data from SSE

```python
datafile = os.path.abspath("../data/sse_data/He_star.dat")
names = ["mass","mass_he","time"]
He_star = np.genfromtxt(datafile, dtype=None, usecols=range(len(names)), names=names)
```

### Let's compare an analytic estimate for the He-core mass with results from SSE

```python
# Analytic estimate: M_He,core = M_0 * M^(1/xi)
m_i = np.arange(1.0,40.0, 0.01)
M_0 = 0.073
xi = 0.704
m_He_core = M_0 * np.power(m_i,1.0/xi)


# Now plot two lines
plt.plot(m_i, m_He_core, color='k', linestyle='--', label="Analytic")
plt.plot(He_star["mass"], He_star["mass_he"], color='k', linestyle='-', label='SSE')

plt.xlabel(r"Initial mass (M$_{\odot}$)")
plt.ylabel("He core mass (M$_{\odot}$)")
plt.legend(loc=2)

plt.show()
```

### Now, let's look at the He ignition time

```python
plt.plot(He_star["mass"], He_star["time"], color='k', linestyle='-', label='SSE')

plt.xlim(6.0,40.0)
plt.ylim(0.0,80.0)
plt.xlabel(r"Initial mass (M$_{\odot}$)")
plt.ylabel("He ignition time (Myr)")

plt.show()
```

### Now, let's look at the difference between the two limits for the effective observation time

```python
m_i = np.arange(3.0,39.0, 0.01)
He_time = interp1d(He_star["mass"], He_star["time"])

t_1 = -He_time(m_i)
t_2 = 2.**(-1.0/xi)*He_time(m_i) - He_time(m_i/2.)

plt.plot(m_i, t_1-t_2, color='k')

plt.xlim(5.,40.)
plt.ylim(0.0,100.0)
plt.xlabel(r"Initial mass (M$_{\odot}$)")
plt.ylabel(r"$\Delta$ t$_{\rm He}$  (Myr)")

plt.show()
```

```python

t_1 = -He_time(m_i)
t_2 = 2.**(-1.0/xi)*He_time(m_i) - He_time(m_i/2.)

plt.plot(m_i, -t_1, color='k', linestyle="--", label=r"$M_1 >> M_2$")
plt.plot(m_i, -t_2, color='k', linestyle="-.", label=r"$M_1 \approx M_2$")

plt.ylim(0.0,50.0)
plt.xlabel(r"Initial mass (M$_{\odot}$)")
plt.ylabel(r"t$_b$ - t$_{\rm eff,obs}$ (Myr)")
plt.legend()

plt.show()
```
